```python
%%sql
SELECT * 
FROM cities
LIMIT 10;
```

```python
%%sql 
SELECT COUNT(DISTINCT name) as numero_ciudades
FROM cities;
```

```python
%%sql
SELECT COUNT(DISTINCT country_code) as numero_paises
FROM cities;
```

```python
%%sql
SELECT country_code, count(name) as ciudades
FROM cities
GROUP BY country_code
ORDER BY ciudades DESC;
```

```python
%%sql
SELECT AVG(metroarea_pop) as avg_metroarea_pop, max(metroarea_pop) as max_metroarea_pop, min(metroarea_pop)
FROM cities
WHERE country_code = 'CHN';
```

```python
%%sql
SELECT name, metroarea_pop
FROM cities
WHERE country_code = 'CHN' and metroarea_pop < 16000000
GROUP BY name;
```

# INNER JOIN POR REGIONES

```python
%%sql
SELECT DISTINCT region 
FROM countries;
```

```python
%%sql
SELECT c1.name, city_proper_pop, metroarea_pop, urbanarea_pop, continent, surface_area, indep_year, gov_form, cap_long, cap_lat,
        CASE WHEN indep_year < 1900 THEN 'Anterior a 1900'
             WHEN indep_year <= 1930 THEN 'Entre 1900 y 1930'
             ELSE 'Posterior a 1930' END as indep_year_group
FROM cities as c1
INNER JOIN countries as c2
on c1.country_code = c2.code
WHERE region = 'North America';
```

```python
%%sql
SELECT c1.name, city_proper_pop, metroarea_pop, urbanarea_pop, continent, surface_area, indep_year, gov_form, cap_long, cap_lat,
        CASE WHEN indep_year < 1900 THEN 'Anterior a 1990'
             WHEN indep_year <= 1930 THEN 'Entre 1990 y 1930'
             ELSE 'Posterior a 1930' END as indep_year_group
FROM cities as c1
INNER JOIN countries as c2
on c1.country_code = c2.code
WHERE region = 'Central America';
```
